```python
import pandas as pd

# Шаг 1: Загрузка данных
titles = pd.read_csv('data/titles_file.csv')
keywords = pd.read_csv('data/keywords_file.csv')

# Шаг 2: Обработка keywords
keywords['Keyword'] = keywords['Keyword'].str.split()

# Шаг 3: Группировка keywords по URL
keywords = keywords.explode('Keyword')
keywords = keywords.groupby('URL')['Keyword'].apply(lambda x: ', '.join(x)).reset_index()

# Шаг 4: Слияние titles и keywords по URL
result = pd.merge(titles, keywords, on='URL', how='inner')

# Шаг 5: Запись результатов
result.to_csv('data/result.csv', index=False)
```

```python
import pandas as pd
from pymystem3 import Mystem

# Инициализация лемматизатора
m = Mystem()

# Функция для лемматизации текста
def lemmatize(text):
    return [lemma for lemma in m.lemmatize(text) if lemma.strip()]

# Загрузка данных
titles = pd.read_csv('data/titles_file.csv')
keywords = pd.read_csv('data/keywords_file.csv')

# Лемматизация ключевых слов
keywords['Keyword'] = keywords['Keyword'].apply(lambda x: ', '.join(lemmatize(x)))

# Группировка и объединение ключевых слов по URL
keywords_grouped = keywords.groupby('URL')['Keyword'].apply(lambda x: ', '.join(x)).reset_index()

# Объединение двух датафреймов по URL
result = pd.merge(titles, keywords_grouped, on='URL')

# Лемматизация заголовков
result['Title'] = result['Title'].apply(lambda x: ' '.join(lemmatize(x)))

# Проверяем каждое слово из "Keyword" на предмет наличия в "Title"
result['Not_in_Title'] = result.apply(lambda row: [word for word in row['Keyword'].split(', ') if word not in row['Title']], axis=1)

# Сохранение результата в CSV-файл
result.to_csv('data/result.csv', index=False)
```
