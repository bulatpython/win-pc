```python
import pandas as pd

data = pd.read_csv('data/heart.csv')
data.head()
```

```python
heart = data

heart['old'] = heart['age'].apply(lambda x: 1 if x > 60 else 0)

display(heart)

print(heart['old'].sum())
```

```python
trestbps = pd.DataFrame({
    'Возраст (лет)': ['до 20', '21-30', '31-40', '41-50', '51-60', '61 и старше'],
    'Мужчины': ['123/76', '126/79', '129/81', '135/83', '142/85', '142/80'],
    'Женщины': ['116/72', '120/75', '127/80', '137/84', '144/85','159/85'],
})
display(trestbps)
```

```python
def trestbps_mean(*args):
    for x, y in args:
        if x <= 20 and y == 1:
            return 123
        elif 20 < x <= 30 and y == 1:
            return 126
        elif 30 < x <= 40 and y == 1:
            return 129
        elif 40 < x <= 50 and y == 1:
            return 135
        elif 50 < x <= 60 and y == 1:
            return 142
        elif x > 60 and y == 1:
            return 142
        elif x <= 20 and y == 0:
            return 116
        elif 20 < x <= 30 and y == 0:
            return 120
        elif 30 < x <= 40 and y == 0:
            return 127
        elif 40 < x <= 50 and y == 0:
            return 137
        elif 50 < x <= 60 and y == 0:
            return 144
        else:
            return 159

heart['trestbps_mean'] = heart[['age', 'sex']].apply(trestbps_mean, axis = 1)
```
